```python
import pandas as pd
import numpy as  np
data=pd.read_excel("slr08.xls")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data = scaler.fit_transform(data)
```

```python
x=data[:,0]
y=data[:,1]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.1,random_state=5)
```

```python
m_0=1 
c_0 =-1
eta = 0.1
del_m = 1
del_c = 1
max_iters = 10000
iters = 0
mine=[]
avge=[]
def deriv(m,c,x,y):
    m_dev=0
    c_dev=0
    for j in range(len(x)):
        m_dev=m_dev-(1*(y[j]-m*x[j]-c)*x[j])
        c_dev=c_dev-(1*(y[j]-m*x[j]-c))
    m_dev=m_dev/len(x)
    c_dev=c_dev/len(x)
    
    return m_dev,c_dev

while iters<max_iters:
    
    for i in range(len(xtrain)):
        prev_m = m_0
        prev_c = c_0
        if iters==999:
            mine.append(((ytrain[i]-m_0*xtrain[i]-c_0)**2)/2)
            avge.append((ytrain[i]-m_0*xtrain[i]-c_0))
        
        del_m,del_c = deriv(prev_m,prev_c,xtrain,ytrain)
        del_m = -eta *del_m
        del_c = -eta *del_c
        m_0 = m_0+del_m
        c_0 = c_0+del_c
      
    iters+=1
    print("Iteration",iters,"\nmvalue is",m_0,"\nc value is",c_0)
print("The local minimum occurs at",m_0,c_0)
```
